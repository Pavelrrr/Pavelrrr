```python
def find_high_correlations(corr_matrix, threshold=0.92):
    high_corr = {}
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                high_corr[colname] = high_corr.get(colname, []) + [corr_matrix.columns[j]]
    return high_corr

high_correlations = find_high_correlations(data.corr())
print("Сильно коррелирующие признаки:")
for key, value in high_correlations.items():
    print(f"{key}: {', '.join(value)}")
```

```python
def remove_correlated_features(dataset, threshold=0.9):
    corr_matrix = dataset.corr()
    high_corr = find_high_correlations(corr_matrix, threshold)
    to_drop = set()
    for key, value in high_corr.items():
        if key not in to_drop:
            to_drop.update(value)
    return dataset.drop(columns=to_drop)

data_reduced = remove_correlated_features(data)
print(f"Количество признаков до удаления: {data.shape[1]}")
print(f"Количество признаков после удаления: {data_reduced.shape[1]}")
```

```python

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure(figsize=(10, 6))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Количество примеров для обучения")
    plt.ylabel("Точность")

    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Точность на обучающей выборке")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Точность на тестовой выборке")

    plt.legend(loc="best")
    return plt
```
